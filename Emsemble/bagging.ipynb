```python
cancer = pd.read_csv("C:\\Users\\DAI.STUDENTSDC\\Desktop\\Machine_Learning\\Cases\\Wisconsin\\BreastCancer.csv", index_col = 0)
# cancer.set_index('ID', inplace = True)
X = cancer.drop('Class', axis = 1)
y = cancer['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 24, test_size = 0.3, stratify = y)
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

# Set up cross-validation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)

# Define parameter grids for each estimator
param_grids = {
    'SVM': {
        'estimator': [SVC()],
        'n_estimators': [10, 50, 100],
        'estimator__C': [0.1, 1, 10]  # Example SVM hyperparameters
    },
    'LogisticRegression': {
        'estimator': [LogisticRegression()],
        'n_estimators': [10, 50, 100],
        'estimator__C': [0.1, 1, 10]  # Example Logistic Regression hyperparameters
    },
    'DecisionTree': {
        'estimator': [DecisionTreeClassifier()],
        'n_estimators': [10, 50, 100],
        'estimator__max_depth': [None, 5, 10]  # Example Decision Tree hyperparameters
    }
}

# Initialize GridSearchCV for each estimator type
search_results = {}
for name, params in param_grids.items():
    bagging_model = BaggingClassifier(random_state=24)
    gcv = GridSearchCV(bagging_model, param_grid=params, scoring='neg_log_loss', cv=kfold, verbose=3)
    gcv.fit(X_train, y_train)
    search_results[name] = [gcv.best_params_, gcv.best_score_]

# Display best parameters for each estimator
print(search_results)
```
