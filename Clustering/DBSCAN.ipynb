## using usa dataset

```python
usa = pd.read_csv("C:\\Users\\DAI.STUDENTSDC\\Desktop\\Machine_Learning\\Datasets\\USArrests.csv", index_col = 0)
usa_scl = StandardScaler().set_output(transform = 'pandas').fit_transform(usa)
```

```python
eps_range = [0.2, 0.4, 0.6, 0.8, 1]
mp_range =  [2, 3, 4, 5]
cnt = 0
a = []
for i in eps_range:
    for j in mp_range:
        clust = DBSCAN(eps = i, min_samples = j)
        clust.fit(usa_scl)
        usa_scl['clust'] = clust.labels_
        if len(set(usa_scl['clust'])) > 2:
            cnt += 1
            clusters = usa_scl[usa_scl['clust'] != -1]
            score = silhouette_score(clusters.iloc[:,:-1], clusters['clust'])
            a.append([cnt, i, j, score])
pa = pd.DataFrame(a, columns = ['Sr', 'eps', 'min_pts', 'sil'])
print("best parameters")
pa.iloc[pa['sil'].idxmax()]
```

## nutrints dataset

```python
nutr = pd.read_csv("C:\\Users\\DAI.STUDENTSDC\\Desktop\\Machine_Learning\\Datasets\\nutrient.csv", index_col = 0)
nutr.head()
```

```python
nutr_scl = StandardScaler().set_output(transform = 'pandas').fit_transform(nutr)
```

```python
eps_range = [0.2, 0.4, 0.6, 0.8, 1, 1.2]
m_range = [2, 3, 4, 5,6 ]
a = []
for i in eps_range:
    for j in m_range:
        clust = DBSCAN(eps = i, min_samples = j)
        clust.fit(nutr_scl)
        nutr_scl['clust'] = clust.labels_
        if len(set(nutr_scl['clust'])) > 2:
            clusters = nutr_scl[nutr_scl['clust'] != -1]
            score = silhouette_score(clusters.iloc[:, :-1], clusters['clust'])
            a.append([i, j, score]) 
pa = pd.DataFrame(a, columns = ['eps', 'min_samples', 'score'])
print("the maximum score is at parameters")
pa.loc[pa['score'].idxmax()]
```
