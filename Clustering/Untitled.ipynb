```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import numpy as np


df = pd.read_csv('/content/dataset.csv')

X = df['title']  
y = df['category'] 


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)


nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)


y_pred = nb.predict(X_test_tfidf)


accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)


print(f"Accuracy: {accuracy}")
print("\nClassification:")
print(report)










































import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Step 1: Load the data
data = pd.read_csv('//content/Articel_datasetbn1.csv')
print("First 5 rows of the dataset:")
print(data.head())

# Step 2: Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(data['বিষয়বস্তু'], data['লেবেল'], test_size=0.3, random_state=42)

# Step 3: Convert text to numerical features using CountVectorizer
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Check shape of the transformed data to make sure it's not empty
print("Shape of X_train_vec:", X_train_vec.shape)
print("Shape of X_test_vec:", X_test_vec.shape)

# Step 4: Train the Naive Bayes model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Step 5: Make predictions
y_pred = model.predict(X_test_vec)

# Step 6: Evaluate the model
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nAccuracy Score:")
print("Accuracy:", accuracy_score(y_test, y_pred))

# Optionally, you can check model's performance on some sample test cases
sample_texts = [
    "লেব্রন জেমস থ্রিলিং গেমে সর্বকালের স্কোরিং রেকর্ড ভেঙে ফেললেন",
    "বিশ্বকাপ ওপেনারে ইউএসডব্লিউএনটি ডমিনেট করেছে"
]

# Vectorize the sample texts using the same vectorizer
sample_vec = vectorizer.transform(sample_texts)

# Predict the labels for sample texts
sample_predictions = model.predict(sample_vec)
print("\nPredictions for Sample Texts:")
for text, prediction in zip(sample_texts, sample_predictions):
    print(f"Text: {text}\nPredicted Label: {prediction}\n")


```
